```python
#TEXT FILE – WRITE & READ
```

```python
#Write text to a file and read it back
with open("demo.txt", "w") as f:
    f.write("Hello Python\n")
    f.write("File handling practice\n")

with open("demo.txt", "r") as f:
    print(f.read())
```

```python
#Read a file line by line
with open("demo.txt", "r") as f:
    for line in f:
        print(line.strip())
```

```python
#Read first 15 characters from a file
with open("demo.txt", "r") as f:
    print(f.read(15))
```

```python
#Read file using readline()
with open("demo.txt", "r") as f:
    print(f.readline().strip())
    print(f.readline().strip())
```

```python
#Append text to a file
with open("demo.txt", "a") as f:
    f.write("This line is appended\n")
```

```python
#Read file after append
with open("demo.txt", "r") as f:
    print(f.read())
```

```python
#Store file lines into a list
with open("demo.txt", "r") as f:
    lines = f.readlines()
    print(lines)
```

```python
#Print lines using list index
print(lines[0].strip())
print(lines[1].strip())
```

```python
#FILE COPY
#Copy content from one file to another
with open("demo.txt", "r") as src, open("copy_demo.txt", "w") as dest:
    for line in src:
        dest.write(line)
```

```python
#Read copied file
with open("copy_demo.txt", "r") as f:
    print(f.read())
```

```python
#SEEK & R+
#Modify file using seek()
with open("demo.txt", "r+") as f:
    f.seek(6)
    f.write("JAVA")
```

```python
#Read updated content
with open("demo.txt", "r") as f:
    print(f.read())
```

```python
#EXCEL FILES – CREATE & WRITE
```

```python
#Create DataFrame and write to Excel
import pandas as pd

data = {
    "Name": ["Tom", "Jerry", "Spike"],
    "Marks": [75, 88, 60]
}

df = pd.DataFrame(data)
df.to_excel("marks.xlsx", index=False)
```

```python
#
df = pd.read_excel("marks.xlsx")
print(df)
```

```python
#EXCEL – BASIC OPERATIONS
```

```python
#Display students with marks > 70
print(df[df["Marks"] > 70])
```

```python
#Find average marks
print("Average:", df["Marks"].mean())
```

```python
#count number of lines in file
count = 0
with open("demo.txt", "r") as f:
    for _ in f:
        count += 1

print("Total lines:", count)
```

```python
#Check if a word exists in file
with open("demo.txt", "r") as f:
    content = f.read()

if "Python" in content:
    print("Word found")
else:
    print("Word not found")
```
