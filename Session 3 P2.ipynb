```python
#Iterator : An iterator is an object that can be looped over
#An object that allows us to iterate over a sequence
#function:iter()->returns iterator item
#next()->returns next item
```

```python
nums = [5, 6, 3, 6, 10]
it = iter(nums)

print(next(it))  
print(next(it))  
print(next(it))  
print(next(it))
print(next(it))
```

```python
#generator
#This is a special type of iterator/function that produces values one at a time, instead of creating the entire location.
#A generator is a function with 'yield'
#yield keyword gives a pause & resumes where it went left off
#It is used to generate values on the fly, saving memory
```

```python
def my_gen():
    yield 1
    yield 2
    yield 3

g = my_gen()
print(next(g))  # 1
print(next(g))
print(next(g))
```

```python
#lambda
#It is a Small and anonymous function.
#Syntax: lambda arguments: expression
```

```python
square = lambda x: x * x
print(square(5))  
```

```python
#map
#Applies a function to each element of an iterable.
nums = [5, 6, 3, 6, 10]
out = list(map(lambda x: x * 2, nums))
print(out)  
```

```python
c = [35, 10, 20, 30]

f = list(map(lambda x: x * 9/5 + 32, c))
print(f)
```

```python
numbers=[1,2,3,4,5]
def square(num):
    return num*num
squared_numbers = list(map(square, numbers))
print(squared_numbers) 
```

```python
def to_celsius(f):
    return (f - 32) * 5/9

f = [32, 50, 68]
c = list(map(to_celsius, f))
print(c)
```

```python
#filter
#Filters elements based on True/False condition.
nums = [5, 6, 3, 6, 10]
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)  
```

```python
nums = [10, 55, 23, 90, 60]

big = list(filter(lambda x: x > 50, nums))
print(big)
```

```python
#reduce()
#Reduces a list into one value.
#Must be imported from functools.
```

```python
from functools import reduce
nums = [5, 6, 3, 6, 10]
result = reduce(lambda x, y: x + y, nums)
print(result)  
```

```python
f = [32, 50, 68, 86]

c = list(map(lambda x: (x - 32) * 5/9, f))
print(c)
```

```python
from functools import reduce
numbers =  [1,2,3,4,5]
def add(x,y):
     return x + y
total = reduce(add, numbers)
print(total)
```

```python
#without lambda
f = [32, 50, 68, 86]
c = [(x - 32) * 5/9 for x in f]
print(c)
```
