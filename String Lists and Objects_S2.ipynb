```python
#STRINGS
#What is a String?
#A string is a sequence of characters
#Strings are immutable (cannot be changed after creation)
#Indexing starts from 0
```

```python
#STRING INDEXING & SLICING
```

```python
#Extract the middle character(s) of a string
s = "Python"
mid = len(s) // 2
print(s[mid])          # odd length
```

```python
#Reverse words in a sentence (not characters)
s = "Python is powerful"
words = s.split()
print(" ".join(words[::-1]))
```

```python
#STRING IMMUTABILITY & TRANSFORMATION
```

```python
#Replace character at index 2 with 'X'
s = "coding"
s = s[:2] + "X" + s[3:]
print(s)
```

```python
#Remove all occurrences of 'a'
s = "data analytics"
print(s.replace('a', ''))
```

```python
#STRING SEARCHING & COUNTING
```

```python
#First non-repeating character
s = "swiss"
for ch in s:
    if s.count(ch) == 1:
        print(ch)
        break
```

```python
#Count frequency of each character
s = "hello"
freq = {}

for ch in s:
    freq[ch] = freq.get(ch, 0) + 1

print(freq)
```

```python
#STRING VALIDATION & CHECKS
```

```python
#Check palindrome (ignoring spaces & case)
s = "Never Odd Or Even"
clean = s.replace(" ", "").lower()
print(clean == clean[::-1])
```

```python
#Check if two strings are anagrams
s1 = "listen"
s2 = "silent"
print(sorted(s1) == sorted(s2))
```

```python
#LIST INDEXING & SLICING
```

```python
#Rotate list right by one
lst = [1,2,3,4,5]
rotated = [lst[-1]] + lst[:-1]
print(rotated)
```

```python
#Extract every third element
lst = [1,2,3,4,5,6,7,8,9]
print(lst[::3])
```

```python
#LIST MUTABILITY & IN-PLACE OPERATIONS
```

```python
#Replace negatives with zero
lst = [10, -5, 3, -2]
for i in range(len(lst)):
    if lst[i] < 0:
        lst[i] = 0

print(lst)
```

```python
#Remove duplicates (Ordering)
lst = [1,2,2,3,1,4]
unique = []

for x in lst:
    if x not in unique:
        unique.append(x)

print(unique)
```

```python
#LIST SEARCHING & AGGREGATION
```

```python
#Find second largest number
lst = [10, 40, 30, 20]
lst = list(set(lst))
lst.sort()
print(lst[-2])
```

```python
#Find all pairs with sum = 6
lst = [1,2,3,4,5]
pairs = []

for i in range(len(lst)):
    for j in range(i+1, len(lst)):
        if lst[i] + lst[j] == 6:
            pairs.append((lst[i], lst[j]))

print(pairs)
```

```python
#NESTED LISTS / MATRICES
```

```python
#Flatten a nested list
matrix = [[1,2],[3,4],[5,6]]
flat = []

for row in matrix:
    for num in row:
        flat.append(num)

print(flat)
```

```python
#Extract diagonal elements
matrix = [[1,2,3],[4,5,6],[7,8,9]]
diag = []

for i in range(len(matrix)):
    diag.append(matrix[i][i])

print(diag)
```

```python
#LIST COMPREHENSIONS
evens = [x for x in range(1, 21) if x % 2 == 0]
print(evens)
```

```python
#Flatten matrix using comprehension
matrix = [[1,2],[3,4],[5,6]]
flat = [num for row in matrix for num in row]
print(flat)
```

```python
#STRING + LIST COMBINED
```

```python
#Find longest word in sentence
s = "Python makes coding enjoyable"
words = s.split()
print(max(words, key=len))
```

```python
#Word frequency
s = "python is easy and python is powerful"
words = s.split()
freq = {}

for w in words:
    freq[w] = freq.get(w, 0) + 1

print(freq)
```
