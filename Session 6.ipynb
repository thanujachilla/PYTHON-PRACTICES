```python
#Write a program that uses the math module to check whether a number is a perfect square.
import math

num = 36
root = math.sqrt(num)

if root.is_integer():
    print("Perfect square")
else:
    print("Not a perfect square")
```

```python
#Insert multiple rows using a loop.
import sqlite3

conn = sqlite3.connect("marks.db")
conn.execute("create table if not exists student(id int, marks int)")

data = [(1, 80), (2, 90), (3, 70)]

for row in data:
    conn.execute("insert into student values (?, ?)", row)

conn.commit()
conn.close()
```

```python
#Fetch only students with marks > 75.
import sqlite3

conn = sqlite3.connect("marks.db")
cursor = conn.execute("select * from student where marks > 75")

for row in cursor:
    print(row)

conn.close()
```

```python
#Sample data
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "Alice", "Bob"],
    "Subject": ["Math", "Math", "Science", "Science", "Science"],
    "Marks": [85, 90, 78, 88, 95]
}

df = pd.DataFrame(data)
```

```python
#How do you filter rows where Marks are greater than 85?
df[df["Marks"] > 85]
```

```python
#How do you display only records where Subject is "Science"?
df[df["Subject"] == "Science"]
```

```python
#How do you sort the DataFrame by Marks in descending order?
df.sort_values(by="Marks", ascending=False)
```

```python
#How do you sort by Subject first and then by Marks?
df.sort_values(by=["Subject", "Marks"])
```

```python
#How do you find the average marks for each subject?
df.groupby("Subject")["Marks"].mean()
```

```python
#How do you count number of students per subject?
df.groupby("Subject")["Name"].count()
```

```python
#How do you add a column Result (Pass if Marks ≥ 80 else Fail)?
df["Result"] = df["Marks"].apply(lambda x: "Pass" if x >= 80 else "Fail")
#apply() runs logic on each value in the column and creates a new column.
```

```python
#How do you add a column that gives 5 bonus marks to everyone?
df["Bonus_Marks"] = df["Marks"] + 5
#Pandas supports vectorized operations, so the addition applies to all rows at once.
```

```python
#How do you identify duplicate student names?
df[df.duplicated("Name")]
```

```python
#How do you remove duplicate names but keep the first occurrence?
df.drop_duplicates(subset="Name")
```

```python
#How do you rename the column "Marks" to "Score"?
df.rename(columns={"Marks": "Score"})
```

```python
#How do you select only Name and Marks columns?
df[["Name", "Marks"]]
```

```python
#How do you label students as "Topper" if Marks ≥ 90?
df["Category"] = df["Marks"].apply(
    lambda x: "Topper" if x >= 90 else "Regular"
)
```

```python
#How do you increase Marks by 10% only for Science students?
df.loc[df["Subject"] == "Science", "Marks"] *= 1.10
```

```python
#How do you find maximum and minimum marks?
df["Marks"].max(), df["Marks"].min()
```

```python
#How do you find the overall average marks?
df["Marks"].mean()
```

```python
#How do you find average marks of students who passed (Marks ≥ 80)?
df[df["Marks"] >= 80]["Marks"].mean()
#First filters passing students, then computes average.
```
